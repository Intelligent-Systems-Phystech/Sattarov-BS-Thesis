```python

def nlp_pipeline(text):

    text = text.lower()
    text = text.replace('\n', ' ').replace('\r', '')
    text = ' '.join(text.split())
    text = re.sub(r"[A-Za-z\.]*[0-9]+[A-Za-z%°\.]*", "", text)
    text = re.sub(r"(\s\-\s|-$)", "", text)
    text = re.sub(r"[,\!\?\%\(\)\/\"]", "", text)
    text = re.sub(r"\&\S*\s", "", text)
    text = re.sub(r"\&", "", text)
    text = re.sub(r"\+", "", text)
    text = re.sub(r"\#", "", text)
    text = re.sub(r"\$", "", text)
    text = re.sub(r"\£", "", text)
    text = re.sub(r"\%", "", text)
    text = re.sub(r"\:", "", text)
    text = re.sub(r"\@", "", text)
    text = re.sub(r"\-", "", text)

    return text
```

```python
text = pd.read_csv("result with speaker.csv")
number_of_texts = text["number_of_text"]
number_of_texts = number_of_texts.astype('str')
speakers = text[" speaker"]
corpus = text[" text"]
corpus = corpus.astype('str')
corpus_clean = corpus.apply(nlp_pipeline)
corpus_clean.head()
length = corpus.size
print(length)
```

```python
dict_speakers ={
    
}
n_phrases = {
    
}

neg_polarity = []#only! negative polarities 
pos_polarity = []#only! positive polarities 
polarity = []
extreme = []
writing = open("extremes.txt", "w", encoding="utf-8")
#for tweet in corpus_clean:
for i in range(length):
    tweet = corpus_clean[i]
    speaker = speakers[i]
    PolarityUnit = TextBlob(tweet,pos_tagger=PatternTagger(),analyzer=PatternAnalyzer()).sentiment[0]
    if dict_speakers.get(speaker) == None:
        dict_speakers[speaker] = PolarityUnit
        n_phrases[speaker] = 1
    else:
        dict_speakers[speaker] = (dict_speakers[speaker]*n_phrases[speaker]+PolarityUnit)/(n_phrases[speaker]+1)
        n_phrases[speaker] += 1
    if(abs(PolarityUnit)>0.8):
        writing.write(tweet)
        writing.write('\n' )
        writing.write(str(TextBlob(tweet,pos_tagger=PatternTagger(),analyzer=PatternAnalyzer()).sentiment[0]))
        writing.write('\n' )
    if PolarityUnit < 0:
        neg_polarity.append(PolarityUnit)
    else:
        neg_polarity.append(0)
    polarity.append(PolarityUnit)
        
```
